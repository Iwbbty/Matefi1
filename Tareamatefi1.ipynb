```python
# Código en Python para mostrar los resultados de cada problema

def resolver_problemas():
    # Problema 1: Un capital es impuesto al 50%. ¿En cuántos meses produce el 25% del monto?
    def problema_1():
        r_mensual = 0.50 / 12
        interes_deseado = 0.25  # 25% del capital
        t = interes_deseado / r_mensual
        return t

    # Problema 2: Interés de S/. 120,000 durante 2 meses y 10 días al 12% trimestral.
    def problema_2():
        C = 120000
        r_trimestral = 0.12
        tiempo_meses = 2 + 10 / 30
        r_mensual = r_trimestral / 3
        interes = C * r_mensual * tiempo_meses
        return interes

    # Problema 3: Interés de S/. 6,000 al 0.5% mensual durante 2 años, 8 meses y 8 días.
    def problema_3():
        C = 6000
        r_mensual = 0.005
        tiempo_total_meses = (2 * 12) + 8 + (8 / 30)
        interes = C * r_mensual * tiempo_total_meses
        return interes

    # Problema 4: Calcular la tasa necesaria para el resto del capital.
    def problema_4():
        r_total = 0.11  # 11% anual total
        r_parte1 = 0.09  # 9% para la primera parte del capital (1/3)
        r_parte2 = (r_total - (1/3 * r_parte1)) / (2/3)
        return r_parte2

    # Problema 5: Determinar el capital depositado si después de 6 trimestres se generaron S/. 9,288.
    def problema_5():
        M = 9288
        r_bimestral = 0.08
        tiempo_trimestres = 6
        capital = M / (1 + r_bimestral * tiempo_trimestres)
        return capital

    # Problema 6: Capital necesario si al 4% genera 14,400 soles.
    def problema_6():
        interes_extra = 2400
        r1 = 0.04
        r2 = 0.05
        capital = interes_extra / (r2 - r1)
        return capital

    # Problema 7: Interés generado por un capital de S/. 30,000 al 1% anual durante 6 meses.
    def problema_7():
        C = 30000
        r_anual = 0.01
        tiempo_meses = 6 / 12
        interes = C * r_anual * tiempo_meses
        return interes

    # Problema 8: Interés generado por un capital de S/. 80,000 del 20 de marzo al 15 de abril al 18% anual.
    def problema_8():
        C = 80000
        r_anual = 0.18
        tiempo_dias = 26 / 360
        interes = C * r_anual * tiempo_dias
        return interes

    # Problema 9: Mayor capital en el banco con rédito del 5% semestral.
    def problema_9():
        menor_monto = 4720
        r_trimestral = 0.04
        r_semestral = 0.05
        capital_trimestral = menor_monto / (r_trimestral * 4)
        capital_semestral = capital_trimestral * (r_trimestral / r_semestral)
        return capital_semestral

    # Problema 10: Tasa cuatrimestral impuesta por Victoria para obtener un extra de S/. 480 en 10 meses.
    def problema_10():
        monto_8_meses = 25920
        diferencia_interes = 480
        tiempo_adicional = 2 / 4
        r_cuatrimestral = diferencia_interes / (monto_8_meses * tiempo_adicional)
        return r_cuatrimestral * 100

    # Ejecución de los problemas
    return {
        "Problema 1": problema_1(),
        "Problema 2": problema_2(),
        "Problema 3": problema_3(),
        "Problema 4": problema_4() * 100,  # Convertimos a porcentaje
        "Problema 5": problema_5(),
        "Problema 6": problema_6(),
        "Problema 7": problema_7(),
        "Problema 8": problema_8(),
        "Problema 9": problema_9(),
        "Problema 10": problema_10()
    }

# Ejecutamos y mostramos resultados
resultados = resolver_problemas()
for problema, resultado in resultados.items():
    print(f"{problema}: {resultado}")

```
